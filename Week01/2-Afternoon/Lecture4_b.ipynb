## Symbolic Computing

Let us plot a figure with its derivatives:

```python
# Step 1: Declare x as a symbol
x = sympy.Symbol("x")
```

```python
# Step 2: Declare our function
f = x ** 3 + 5 * x ** 2 + 10
f
```

```python
# First derivative
f.diff(x)
```

```python
# Second derivative
f.diff(x, 2)
```

## Symbolic Computing

Let us plot the equation:

```python
# Plotting figure
p = plot(f, f.diff(x), f.diff(x, 2), 
         xlim = (-5, 2), 
         ylim = (-20, 40),
         xlabel = '',
         ylabel = '', 
         legend = True)
```

## Symbolic Computing

```python
a, b, x, y = sympy.symbols("a, b, x, y")
f = sympy.Function("f")(x)
f
```

What is a solution? We characterize it as $x$ such that $f(x) = 0$.

A little theorem to help us up:

**Theorem**: If $f \in C[a, b]$ (meaning: $f$ is continuous in the interval $[a, b]$) and $K$ is a number between $f(a)$ and $f(b)$, then there exists $c \in (a, b)$ for which $f(c) = K$
